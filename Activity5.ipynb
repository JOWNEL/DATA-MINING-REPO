```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
net=pd.read_csv('datasets/Internet-Demographic.csv')
```

```python
net
```

<h1>
    1. Distribution of Birth Rates </h1>

```python
plt.figure(figsize=(10,6))


sns.histplot(net["Birth rate"], bins=20, kde=True, color="skyblue")


plt.title("Distribution of Birth Rates", fontsize=14)
plt.xlabel("Birth Rate (per 1,000 people)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.show()
```

<h1>
   2. Top 5 Countries with the Highest Internet Usage
</h1>

```python
top_5_internet = net.sort_values(by='Internet users', ascending=False).head(5)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Internet users', y='Country Name', data=top_5_internet, palette='viridis')
plt.title('Top 5 Countries with the Highest Internet Usage', fontsize=16)
plt.xlabel('Internet Users (%)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

<h1>
    3. Count of Countries by Income Group</h1>

```python
income_group_counts = net['Income Group'].value_counts()


print(income_group_counts)

plt.figure(figsize=(10, 6))
income_group_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Count of Countries by Income Group', fontsize=16)
plt.xlabel('Income Group', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h1>
    4. Box Plot of Internet Usage </h1>

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=net['Internet users'], color='lightblue')
plt.title('Distribution of Internet Usage Across Countries', fontsize=16)
plt.xlabel('Internet Users (%)', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.show()
```

<h1>
   5. Countries with No Internet Access
</h1>

```python
low_internet_countries = net[net['Internet users'] < 1]

print(low_internet_countries[['Country Name', 'Internet users']])


plt.figure(figsize=(10, 6))
sns.barplot(x='Internet users', y='Country Name', data=low_internet_countries, palette='Reds_r')
plt.title('Countries with Extremely Low Internet Usage (< 1%)', fontsize=16)
plt.xlabel('Internet Users (%)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.show()
```

<h1>
   6. Average Birth Rate
 </h1>

```python

average_birth_rate = net['Birth rate'].mean()


print(f"Average Birth Rate Across All Countries: {average_birth_rate:.2f} per 1,000 people")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(net['Birth rate'], bins=30, kde=True, color='skyblue', alpha=0.7)


plt.axvline(average_birth_rate, color='red', linestyle='--', linewidth=2, label=f'Average Birth Rate: {average_birth_rate:.2f}')

plt.title('Distribution of Birth Rates Across Countries', fontsize=16)
plt.xlabel('Birth Rate (per 1,000 people)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

```

<h1>
   7. Scatter Plot: Birth Rate vs. Internet Usage
 </h1>

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Birth rate', y='Internet users', data=net, hue='Income Group', palette='Set1', s=100)


plt.title('Scatter Plot: Birth Rate vs. Internet Usage', fontsize=16)
plt.xlabel('Birth Rate (per 1,000 people)', fontsize=12)
plt.ylabel('Internet Users (%)', fontsize=12)


plt.grid(axis='both', linestyle='--', alpha=0.7)


plt.legend(title='Income Group', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.tight_layout()
plt.show()
```

<h1>
    8. Countries with Birth Rates Above 40
</h1>

```python

high_birth_rate_countries = net[net['Birth rate'] > 40]


print(high_birth_rate_countries[['Country Name', 'Birth rate']])
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Birth rate', y='Country Name', data=high_birth_rate_countries, palette='Reds_r')
plt.title('Countries with Birth Rates Above 40 (per 1,000 people)', fontsize=16)
plt.xlabel('Birth Rate (per 1,000 people)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

<h1>
   9. Pie Chart of Income Groups</h1>

```python

income_group_counts = net['Income Group'].value_counts()
```

```python
income_group_counts = net['Income Group'].value_counts()


labels = income_group_counts.index
sizes = income_group_counts.values


colors = ['skyblue', 'lightgreen', 'salmon', 'gold']


plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)


plt.title('Proportion of Countries by Income Group', fontsize=16)


plt.legend(title='Income Group', loc='upper right')


plt.show()
```

<h1>
  10. Correlatioon between birth rate and internet usage</h1>

```python

correlation = net[['Birth rate', 'Internet users']].corr(method='pearson')


print(correlation)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Birth rate', y='Internet users', data=net, hue='Income Group', palette='Set1', s=100)

# Add regression line
sns.regplot(x='Birth rate', y='Internet users', data=net, scatter=False, color='red', label='Regression Line')

# Add title and axis labels
plt.title('Scatter Plot: Birth Rate vs. Internet Usage', fontsize=16)
plt.xlabel('Birth Rate (per 1,000 people)', fontsize=12)
plt.ylabel('Internet Users (%)', fontsize=12)

# Add gridlines
plt.grid(axis='both', linestyle='--', alpha=0.7)

# Add legend
plt.legend(title='Income Group', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
