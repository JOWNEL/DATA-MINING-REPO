```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')

import folium 
from folium.plugins import HeatMap
```

```python
terror = pd.read_csv("datasets//Global_terrorist.csv", encoding = 'ISO-8859-1')
```

```python
ter =terror.columns.to_list()

output_filename = "my_data.txt"

try:
    with open(output_filename, 'w') as file:
        for item in ter:
            file.write(item + '\n')  
    print(f"Data successfully written to '{output_filename}'")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def normalizer(df,column):
    normalized  = (df[column] - df[column].min()) / \
    (df[column].max() - df[column].min())
    return normalized
```

<h1> INSIGHT 1: Educational Institution Attacks in MENA( Beofre in 2000)</h1>

```python
filtered_df_education = terror[
    (terror['region_txt'] == 'Middle East & North Africa') &
    (terror['iyear'] <= 2000) &
    (terror['targtype1_txt'] == 'Educational Institution')
]
print(f'Filtered dataset size: {len(filtered_df_education)} rows')
```

```python
aggregated_df_edu = filtered_df_education.groupby(['latitude','longitude']).size().reset_index(name='incident_count')


def normalizer(df, col):
    return (df[col] - df[col].min()) / (df[col].max() - df[col].min())

aggregated_df_edu['normalized_weight'] = normalizer(aggregated_df_edu, 'incident_count')


heat_data = aggregated_df_edu[['latitude','longitude','normalized_weight']].values.tolist()


m = folium.Map(location=[30,40], zoom_start=4)
HeatMap(heat_data).add_to(m)


m
```

<h1>“A significant number of terrorist attacks against educational institutions occurred in Turkey and surrounding areas before 2000, reflecting ongoing regional unrest and targeted attacks on learning environments.”<h1>

<h1> Insight 2: Government Attacks in South Asia (2000–2010)<h1>

```python
filtered_df_gov = terror[
    (terror['region_txt'] == 'South Asia') &
    (terror['iyear'] >= 2000) &
    (terror['iyear'] <= 2010) &
    (terror['targtype1_txt'] == 'Government (General)')
]

print(f'Filtered dataset size: {len(filtered_df_gov)} rows')

# Group by location and count
aggregated_df_gov = filtered_df_gov.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')


# Normalize weights
aggregated_df_gov['normalized_weight'] = normalizer(aggregated_df_gov, 'incident_count')

# Heatmap data format
heat_data_gov = aggregated_df_gov[['latitude', 'longitude', 'normalized_weight']].values.tolist()

# Create map
m2 = folium.Map(location=[28, 75], zoom_start=4)
HeatMap(heat_data_gov).add_to(m2)

# Show heatmap
m2
```

<h2>Between 2000 and 2010, South Asia experienced a high concentration of attacks targeting government institutions, with hotspots in Afghanistan and Pakistan—regions significantly impacted by internal conflict and insurgency.<h2>

<h1> Insight 3: Civilian Attacks in Sub-Saharan Africa (2010–2020)<h1>

```python
filtered_df_civilian = terror[
    (terror['region_txt'] == 'Sub-Saharan Africa') &
    (terror['iyear'] >= 2010) &
    (terror['iyear'] <= 2020) &
    (terror['targtype1_txt'] == 'Private Citizens & Property')
]

print(f'Filtered dataset size: {len(filtered_df_civilian)} rows')

aggregated_df_civilian = filtered_df_civilian.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')


aggregated_df_civilian['normalized_weight'] = normalizer(aggregated_df_civilian, 'incident_count')


heat_data_civilian = aggregated_df_civilian[['latitude', 'longitude', 'normalized_weight']].values.tolist()


m3 = folium.Map(location=[5, 20], zoom_start=4)
HeatMap(heat_data_civilian).add_to(m3)


m3
```

<h1>From 2010 to 2020, Sub-Saharan Africa witnessed a rising number of terrorist attacks targeting civilians, with notable intensity in Nigeria and surrounding countries—largely due to extremist insurgencies<h1>

<h1>Insight 4: Military Attacks in Latin America & the Caribbean (1990–2005)<h1>

```python
filtered_df_military = terror[
    (terror['region_txt'] == 'Latin America & Caribbean') &
    (terror['iyear'] >= 1990) &
    (terror['iyear'] <= 2005) &
    (terror['targtype1_txt'] == 'Military')
]

print(f'Filtered dataset size: {len(filtered_df_military)} rows')


aggregated_df_military = filtered_df_military.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
aggregated_df_military['normalized_weight'] = normalizer(aggregated_df_military, 'incident_count')


heat_data_military = aggregated_df_military[['latitude', 'longitude', 'normalized_weight']].values.tolist()


m4 = folium.Map(location=[-10, -70], zoom_start=4)
HeatMap(heat_data_military).add_to(m4)


m4
```

<h1>Between 1990 and 2005, Latin America and the Caribbean experienced a concentration of attacks on military targets, particularly in Colombia and Peru—regions marked by prolonged armed conflicts and guerrilla warfare.<h1>
